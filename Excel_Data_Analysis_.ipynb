```python
import pandas as pd

data = {
    "Name": ["Asha", "Ravi", "Kiran", "Meena", "Rahul"],
    "Math": [85, 78, 92, 70, 88],
    "Science": [90, 72, 85, 65, 80],
    "English": [75, 88, 82, 95, 70]
}

df = pd.DataFrame(data)
df.to_csv("students_scores.csv", index=False)
df
```

```python
df = pd.read_csv("students_scores.csv")
df.head()
```

```python
df["Average"] = df[["Math","Science","English"]].mean(axis=1)
df
```

```python
print("Subject means:\n", df[["Math","Science","English"]].mean())
print("\nTop performer:\n", df.loc[df["Average"].idxmax()])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(df["Name"], df["Average"])
plt.xlabel("Student")
plt.ylabel("Average Score")
plt.title("Average Scores of Students")
plt.show()
```
